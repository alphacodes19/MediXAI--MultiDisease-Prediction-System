```python
import os
import urllib.request
import pathlib
```

```python
FILE_PREFIXES = [
    "DEMO", # Demographics
    "BMX",  # Body Measures
    "BPX",  # Blood Pressure
    "GHB",  # Glycohemoglobin
    "GLU",  # Glucose
    "DIQ",  # Diabetes
]
```

```python
CYCLES = {
    'D': '2005-2006',
    'E': '2007-2008',
    'F': '2009-2010',
    'G': '2011-2012',
    'H': '2013-2014',
    'I': '2015-2016',
    'J': '2017-2018',
}
```

```python
def download_cycle_data(cycle_letter, prefixes=FILE_PREFIXES, out_dir_base="data/raw"):
    if cycle_letter not in CYCLES:
        print(f"Error: Cycle {cycle_letter} not defined.")
        return

    years = CYCLES[cycle_letter]
    BASE = f"https://wwwn.cdc.gov/Nchs/Nhanes/{years}/"
    out_dir = pathlib.Path(out_dir_base) / f"{years}_{cycle_letter}"

    os.makedirs(out_dir, exist_ok=True)
    
    print(f"\n--- Starting Download for Cycle {cycle_letter} ({years}) ---")
    
    for prefix in prefixes:
        if cycle_letter == 'P':
            fn = f"{prefix}_{cycle_letter}.XPT" if prefix != "DEMO" else f"P_{prefix}.XPT"
        
        else:
            fn = f"{prefix}_{cycle_letter}.XPT"
        
        url = BASE + fn
        dest = out_dir / fn
        
        print(f"Downloading {url} → {dest}")
        try:
            urllib.request.urlretrieve(url, dest)
            print("✓ done")
        except urllib.error.HTTPError as e:
            print(f"✗ ERROR: Could not download {url}. Status: {e.code} {e.reason}")
    
    print(f"--- Finished Cycle {cycle_letter} ---")
```

```python
if __name__ == "__main__":
    all_cycles = list(CYCLES.keys())
    
    for cycle in all_cycles:
        download_cycle_data(cycle)
        
    print("\n\nAll downloads attempted.")
```
